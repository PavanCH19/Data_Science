Develop a program to implement Principal Component Analysis (PCA) for reducing the dimensionality of
the Iris dataset from 4 features to 2

```python
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
```

```python
data = load_iris(as_frame=True)
df = data.frame
df
```

```python
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(df[data.feature_names])
reduced_data
```

```python
reduced_df = pd.DataFrame(reduced_data, columns=['pc1', 'pc2'])
reduced_df["target"] = data.target
reduced_df
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(reduced_df, x="pc1", y="pc2", hue="target", palette="Set1")
plt.title("scatterplot for reduced iris data")
plt.show()
```
